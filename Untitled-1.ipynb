```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Gamma_L = np.linspace(0.0, 0.90, 100)
alpha = np.array([0.1, 1, 2])  # dB/m
alpha_Neper = alpha*np.log(10)/20
Pin = 1
l = 1

def to_dB(Pin, PL):
    return 10 * np.log10(Pin / PL)

for i in range(len(alpha_Neper)):
    PL = Pin * np.exp(-2 * alpha_Neper[i] * l) * (1 - abs(Gamma_L)**2)
    plt.plot(Gamma_L, to_dB(Pin, PL), label=f'α = {alpha[i]} dB/m')

plt.title("Power Loss In Lossy Transmission Line As a Function of Reflection")
plt.xlabel("Gamma_L")
plt.ylabel("Loss (dB)")
plt.legend()
plt.grid(True)
plt.show()
```
