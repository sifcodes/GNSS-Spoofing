```python
def ca_code(prn: int):
    # G2 tap pairs for PRNs 1–37 (and some extras) vary by source;
    # for PRN 38 we use (2, 9).
    g2_taps = {38: (2, 9)}

    t1, t2 = g2_taps[prn]

    # 10-bit registers, index 1..10 (ignore index 0)
    g1 = [None] + [1]*10
    g2 = [None] + [1]*10

    code = []

    for _ in range(1023):
        g1_out = g1[10]
        g2_out = g2[t1] ^ g2[t2]
        code.append(g1_out ^ g2_out)

        # feedbacks
        g1_fb = g1[3] ^ g1[10]
        g2_fb = g2[2] ^ g2[3] ^ g2[6] ^ g2[8] ^ g2[9] ^ g2[10]

        # shift right (10 -> 9 -> ... -> 1), insert fb at 1
        g1 = [None, g1_fb] + g1[1:10]
        g2 = [None, g2_fb] + g2[1:10]

    return code  # list of 0/1 length 1023

chips = ca_code(38)
print("First 20 chips:", "".join(map(str, chips)))
print("Length:", len(chips))

```
